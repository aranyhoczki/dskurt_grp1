```python
import pandas as pd  # pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
df_train = pd.read_csv(trainset_name, sep=",")
```

```python
df_train.head()
```

```python
## TODO : nan, infiniti -> 0
df_train.replace([np.inf, -np.inf], np.nan, inplace=True)
df_train.fillna(0, inplace=True)
```

```python
# a device_model-t transzformáljuk
```

```python
factor = pd.factorize(df_train['phone_brand_eng'])
df_train.phone_brand_eng = factor[0]
definitions_phone_brand_eng = factor[1]
factor = pd.factorize(df_train['device_model'])
df_train.device_model = factor[0]
definitions_device_model = factor[1]
factor = pd.factorize(df_train['gender'])
df_train.gender = factor[0]
definitions_gender = factor[1]
```

```python
len(df_train.groupby("group").count())
```

```python
magyarazo_valtozok = []
for m in df_train.columns:
    if (m not in ['Unnamed: 0', 'device_id', 'min_timestamp', 'max_timestamp', 'group', 'gender', 'age', 'phone_brand' ]) and (not m.startswith('yyylabel_i_')):
        magyarazo_valtozok += [m]
#magyarazo_valtozok
X_train, X_test, y_train, y_test = train_test_split(df_train[magyarazo_valtozok], df_train['group'], test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
magyarazo_valtozok
```

```python
X_train.head()
```

```python
# normalizálás
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(X_train)  
X_train_scd = pd.DataFrame(scaler.transform(X_train))
X_train_scd.columns = X_train.columns
X_test_scd = pd.DataFrame(scaler.transform(X_test))
X_test_scd.columns = X_test.columns
X_test_scd.head()
```

```python
#Create a svm Classifier
clf = svm.SVC(kernel='rbf') # Linear Kernel rbf/linear
#Train the model using the training sets
clf.fit(X_train_scd, y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test_scd)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
